```
ReImSurface[f_] :=
  ParametricPlot3D[
    {Re[z], Im[z], Re[f[z]]},
    {r, 0.01, 2}, {, 0, 2 Pi},
    PlotPoints -> 100,
    Mesh -> None,
    ColorFunction -> Function[{x, y, z}, ColorData["Rainbow"][/(2 Pi)]],
    ColorFunctionScaling -> False,
    Boxed -> False
  ] /. z -> r Exp[I ]

ReImSurface[Sqrt]
```
