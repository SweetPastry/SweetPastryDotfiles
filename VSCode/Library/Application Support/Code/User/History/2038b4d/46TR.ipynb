```
(* 最简单算式 *)
2+2
```

```
(* 函数 *)
FactorInteger[2^2^5 + 1]    (* Fermet 素数 *)
(* 和以下写法等价 *)
FactorInteger[2^(2^5) + 1]
```

```
(* 虚数运算 *)
(* Pi 和 I 之间没有乘号 *)
E^(Pi I) + 1
```

 `mm` 对整数的位数不做限制, 也就是不区分长短整型等

```
(* 特殊常数之间运算会保留符号 *)
EulerGamma
```

```
(* 数值计算, 50 位精度 *)
N[EulerGamma, 50]   (* N 函数, 任意计算数字精度 *)
```

```
(* 表达式展开函数 *)
Expand[(a+b)^3]
```

```
(* 因式分解函数 *)
Factor[x^3+y^3+z^3-3x y z]    (* 这里没要求 x y z 中间写乘号 *)
(* 空格被解释为乘号 *)
```

```
(* 解方程, 注意使用 == 符号 *)
Solve[x^3 - 2x - 1 == 0, x]
```

```
(* 求极限 *)
Limit[(Tan[x] - x) / (x - Sin[x]), x->0]
```

```
(* 求导数 *)
D[x^x, x]
```

```
(* 求积分 *)
Integrate[x^2Cos[x], x] (* 不定积分是没有积分常数的 *)
```

```
(* 求和函数 *)
Sum[x^(2n) / (n^2 Binomial[2n, n]), {n, Infinity}]   
(* Binomial: 组合数 *)
(* {n, Infinity} 等价于 {n, 1,Infinity} *)
```

```
(* 求微分方程通解 *)
FullSimplify[
    DSolve[y''[x] + y[x] == 8x Sin[x], 
    y[x], x]
]
```

```
(* 求初值问题 *)
FullSimplify[
    DSolve[{
        y''[x] + y[x] == 8x Sin[x], 
        y[0] == 0, 
        y'[0] == 0
},  
        y[x], x]
]
```

```
(* 画图 *)
Plot[
    Sin[1/x],   {x, 0, 1/Pi}
]
```

```
(* 解决左侧毛刺问题 *) 
(* 原理: 将离散点取得密集 *)
Plot[
    Sin[1/x],   {x, 0, 1/Pi},   PlotPoints->10000
]
```
