```matlab
d1 = [8.40, 17.70, 38.00, 58.00, 78.10, 98.30, 117.90, 158.30];
t = [14.16, 17.88, 25.58, 33.50, 41.80, 49.30, 56.90, 73.20];
```

```matlab
p = polyfit(d1, t, 1);
t_fit = polyval(p, d1);
1/p(1)
SS_res = sum((t - t_fit).^2);
SS_tot = sum((t - mean(t)).^2);
R_squared = 1 - SS_res / SS_tot
scatter(d1, t, "filled");
hold on;
box on;
grid on;
plot(d1, t_fit, 'LineWidth', 2);
legend("Experiment Data", "Linear Regression", 'Location', 'northwest');
text(70, 150, 'sound velocity = 353.5(1)m/s', ...
     'FontSize', 12, ...
     'FontWeight', 'bold', ...
     'Color', 'black', ...
     'BackgroundColor', 'white', ...
     'EdgeColor', 'black');
text(50, 350, 'R^2=0.9998', ...
'FontSize', 12, ...
'FontWeight', 'bold', ...
'Color', 'black', ...
'BackgroundColor', 'white', ...
'EdgeColor', 'black');
xlabel('L / mm');
ylabel('t / us')
title('Time Difference Method')
```
