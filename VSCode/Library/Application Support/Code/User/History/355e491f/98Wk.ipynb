```matlab
d1 = [10, 30, 50, 70, 90, 111, 130, 150];
t = [77.20, 132.8, 193.6, 250, 307, 371, 428, 482];
```

```matlab
p = polyfit(d1, t, 1);
t_fit = polyval(p, d1);
1/p(1)
scatter(d1, t, "filled");
hold on;
box on;
grid on;
plot(d1, t_fit, 'LineWidth', 2);
legend("Experiment Data", "Linear Regression", 'Location', 'northwest');
text(70, 150, 'sound velocity = 353.5(1)m/s', ...
     'FontSize', 12, ...
     'FontWeight', 'bold', ...
     'Color', 'black', ...
     'BackgroundColor', 'white', ...
     'EdgeColor', 'black');
xlabel('L/mm');
ylabel('t/us')
title('Time Difference Method')
```

```matlab
n = [1, 2, 3, 4, 5, 6, 7, 8];
d2 = [3.4, 8, 12, 17.4, 22, 25.9, 30.9, 35.7];
Lf = d2 * 37;
p = polyfit(n, Lf, 1);
Lf_fit = polyval(p, n);
2*p(1)
scatter(n, Lf, "filled");
hold on;
box on;
grid on;
plot(n, Lf_fit, 'LineWidth', 2);
legend("Experiment Data", "Linear Regression", 'Location', 'northwest');
text(70, 150, 'sound velocity = 353.5(1)m/s', ...
     'FontSize', 12, ...
     'FontWeight', 'bold', ...
     'Color', 'black', ...
     'BackgroundColor', 'white', ...
     'EdgeColor', 'black');
xlabel('L/mm');
ylabel('t/us')
title('Time Difference Method')
```
