```matlab
p = polyfit(d1, t, 1);
t_fit = polyval(p, d1);
1/p(1)
scatter(d1, t, "filled");
hold on;
box on;
grid on;
plot(d1, t_fit, 'LineWidth', 2);
legend("Experiment Data", "Linear Regression", 'Location', 'northwest');
text(70, 150, 'sound velocity = 353.5(1)m/s', ...
     'FontSize', 12, ...
     'FontWeight', 'bold', ...
     'Color', 'black', ...
     'BackgroundColor', 'white', ...
     'EdgeColor', 'black');
xlabel('L/mm');
ylabel('t/us')
title('Time ')
```
