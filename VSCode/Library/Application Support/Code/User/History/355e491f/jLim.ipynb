```matlab
d1 = [10, 30, 50, 70, 90, 111, 130, 150];
t = [77.20, 132.8, 193.6, 250, 307, 371, 428, 482];
```

```matlab
p = polyfit(d1, t, 1);
t_fit = polyval(p, d1);
scatter(d1, t, "filled");
hold on;
box on;
grid on;
plot(d1, t_fit, 'LineWidth', 2);
legend("Experiment Data", "Linear Regression", 'Location', 'northwest');

xlabel('L/mm');
ylabel('t/us')
```
