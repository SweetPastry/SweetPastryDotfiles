```r
#  ggplot2 
# install.packages("ggplot2")

library(ggplot2)

# 
x <- c(1, 2, 3, 4, 5)
y <- c(2.1, 4.0, 5.8, 8.1, 10.2)
df <- data.frame(x, y)

# 
ggplot(df, aes(x = x, y = y)) +
  geom_point(color = "blue", size = 3) +            # 
  geom_smooth(method = "lm", se = FALSE, color = "red") +  # 
  labs(title = "", x = " x", y = " y") +
  theme_minimal()
```
