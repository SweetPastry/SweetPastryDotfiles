```matlab
data=[randn(2,5),randn(2,3)+ones(2,3)*10];
plot(data(1,:),data(2,:),'bo')
```

```matlab
% 点的数量
numPoints = size(data, 2);

% 计算点之间的欧几里得距离矩阵
distMatrix = zeros(numPoints); % 初始化距离矩阵
for i = 1:numPoints
    for j = i+1:numPoints
        distMatrix(i, j) = norm(data(:, i) - data(:, j)); % 欧几里得距离
        distMatrix(j, i) = distMatrix(i, j); % 对称
    end
end
```
