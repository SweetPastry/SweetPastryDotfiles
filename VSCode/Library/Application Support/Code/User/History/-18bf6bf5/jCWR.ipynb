# 1. `figure(<figure_order>)` 和 `subplot(<row>, <col>, <index>)`

```matlab
% 产生数据
t = linspace(0, 2*pi, 1000)
x = 2*sin(t)
y = 0.5*cos(t) + 0.5
```

```matlab
figure(1)   % 现在在 figure 1 上画图, 缺省是默认采用 figure 1
plot(t, x)
hold on     % 如果不写, 后续的 plot 会覆盖前面的 plot 的内容
plot(t, y)
```

```matlab
% 像这样分别指定两个 figure 就是画“两张”图
figure(1)
plot(t, x)
figure(2)
plot(t, y)
```

```matlab
% 画布上有一行二列两个 figure, 只在第一个上画图, 所以另外一侧是空的
subplot(1, 2, 1)
plot(t, x)
```

```matlab
subplot(1, 2, 1)
plot(t, x)
% 这样就把两个 figure 都画了
subplot(1, 2, 2)
plot(t, y)
```

```matlab
% 横着画
subplot(2, 1, 1)
plot(t, x)
subplot(2, 1, 2)
plot(t, y)
```
