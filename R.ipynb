```r
# 安装并加载 ggplot2 包（第一次用需取消注释）
# install.packages("ggplot2")

library(ggplot2)

# 模拟数据
x <- c(1, 2, 3, 4, 5)
y <- c(2.1, 4.0, 5.8, 8.1, 10.2)
df <- data.frame(x, y)

# 绘图
ggplot(df, aes(x = x, y = y)) +
  geom_point(color = "blue", size = 3) +            # 散点图
  geom_smooth(method = "lm", se = FALSE, color = "red") +  # 拟合直线
  labs(title = "线性拟合图", x = "自变量 x", y = "因变量 y") +
  theme_minimal()
```
